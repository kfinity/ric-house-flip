```python
# save to file
#df.to_csv('data/transfers_2019-2022.csv',index=False)
df = pd.read_csv('data/transfers_2019-2022.csv')
```

```python
# filters
# property classes between 110-150 = single-family houses. include 101 because sometimes 
df = df.loc[df['sale_prclass'].isin([101,110,115,120,130,150])]
```

```python
# houses which changed classification - either built on empty lots, or something like that.
df.loc[df['property_class'] != df['sale_prclass'],['property_class','sale_prclass']]
```

```python
# cleaning
# date from epoch-milliseconds to datetime
df['sale_date'] = pd.to_datetime(df['sale_date'],unit='ms')
```
